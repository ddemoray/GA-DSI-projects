```python
def lnprob(x):
    if x[2] < 0:
        return -np.infty
    else:
        return \
    -0.5*x[2]*np.sum([(e[0] - x[0] + x[1]*np.sin(e[1]))**2 for e in data])\
    +0.5*N*np.log(x[2]) - \
    0.5*(x[0]**2 + x[1]**2) - x[2]
```

```python
nwalkers = 20

ndim = 3
p0 = np.random.rand(nwalkers*ndim).reshape((nwalkers,ndim))
sampler = emcee.EnsembleSampler(nwalkers, ndim, lnprob)
pos, prob, state = sampler.run_mcmc(p0, 1000)
sampler.reset()
pos, prob, state = sampler.run_mcmc(pos, 100000)
samples = sampler.flatchain
```

```python
def MonteCarlo(f, samples):
    N = len(samples)
    return 1/float(N)*sum([f(e) for e in samples])

omega0 = MonteCarlo(lambda x: x[0], samples)
omega1 = MonteCarlo(lambda x: x[1], samples)
lam = MonteCarlo(lambda x: x[2], samples)
print ('o0 = {0}, o1 = {1}, l = {2}'.format(omega0, omega1, lam))
```

```python
Xval = np.arange(0,7)
Reg = [omega0 + np.sin(x) for x in Xval]
#omega0 adjusts Xt of regression
#omega1 does't seem to have an effect on regression
fig, ax = plt.subplots(1,1, figsize = (5,3))
ax.scatter(T,X)
ax.plot(Xval, Reg, label ='Regression', color = 'red')
ax.set_xlabel('t')
ax.set_ylabel('$X_t$')
ax.legend(loc = 'best', fontsize='small')
ax.grid()
```
