```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
```

```python
#import two CSVs from project five into two separate dataframes in pandas.

LoanStatsDF = pd.read_csv('/Users/danieldemoray/GA-DSI-Working/Misc/project-05-assets/LoanStats3d.csv', low_memory=False)
```

```python
LoanStatsDF.head()
```

```python
LoanStatsDF.reset_index(inplace=True )
#reset index to be able to use iloc to pull column names from data set to reset columns
```

```python
# assign variable columns equal to first row of DF
newcols = LoanStatsDF.iloc[0]
LoanStatsDF.rename(columns=newcols, inplace=True)

#remove row 0 from data frame as this has now been used to rename columns
LoanStatsDF.drop(0, inplace=True)
```

```python
LoanStatsDF.head()
```

```python
LoanStatsDF.head()
```

```python
LoanStatsDF['loan_status'] = LoanStatsDF['loan_status'].astype(str)
```

```python
LoanStatsDF['loan_status'].nunique()
```

```python
LoanStatsDF['loan_status'].unique()
```

```python
# Identified Fully Paid and Default loans from data frame in new column
# Completed this to pull out these values.
def calcloan (string):
    if string == 'Fully Paid':
        return 1
    elif string == 'Default':
        return 1
    else:
        return 0

LoanStatsDF['use_these'] = LoanStatsDF['loan_status'].apply(calcloan)
```

```python
newLoanStatsDF = LoanStatsDF.loc[LoanStatsDF['use_these'] == 1]
```

```python
newLoanStatsDF.reset_index(inplace=True)
```

```python
newLoanStatsDF
```

```python
# newLoanStatsDF.loc[newLoanStatsDF['loan_status'] == 'Fully Paid']
```

```python
test = newLoanStatsDF[['loan_status','annual_inc', 'annual_inc_joint', 'dti', 'dti_joint', 'int_rate', \
            'term', 'loan_amnt', 'home_ownership', 'application_type']]
```

```python
print test['annual_inc_joint'].nunique()

print test['dti_joint'].nunique()
# this has all values except default or 
```

```python
def GoodLoan (string):
    if string == 'Fully Paid':
        return 1
    else:
        return 0
test['GoodLoan'] = test['loan_status'].apply(GoodLoan)
```

```python
test = test.drop('loan_status', 1)
```

```python
test['annual_inc'].unique()
```

```python
# renaming test data frame df
df = test
```

```python
df.info()
```

```python
# caste columns at floats
df['annual_inc'] = df['annual_inc'].apply(pd.to_numeric)
df['annual_inc_joint'] = df ['annual_inc_joint'].apply(pd.to_numeric) 
df['dti'] = df ['dti'].apply(pd.to_numeric)
df['dti_joint'] = df ['dti_joint'].apply(pd.to_numeric)
df['loan_amnt'] = df ['loan_amnt'].apply(pd.to_numeric)
```

```python
df.head()
```

```python
df['home_ownership'].unique()
```

```python
df['application_type'].unique()
```

```python
def indiv (string):
    if string == 'INDIVIDUAL':
        return 1
    else:
        return 0
df['Individual_app'] = df['application_type'].apply(indiv)
df = df.drop('application_type', 1)
```

```python
#creating dummy variables for home_ownership column
#reset working data frame from df to df_new to avoid any issues with duplicate columns
#removed home_ownership column after creating dummy variables
df_home = pd.get_dummies(df['home_ownership'])
df_new = pd.concat([df, df_home], axis = 1)
df_new = df_new.drop('home_ownership', 1)
```

```python
#creating dummy variables for term

df_term = pd.get_dummies(df_new['term'])
df2 = pd.concat([df_new, df_term], axis = 1)
df2 = df2.drop('term',1)
df2.rename(columns={' 36 months': 'term_36', ' 60 months': 'term_60'}, inplace=True)
```

```python
#convert %string floats in interest rate to decimal value
def p2f(x):
    return float(x.strip('%'))/100
df2['int_rate'] = df2['int_rate'].apply(p2f)
```

```python
df2.head(10)
```

```python
#removing joint columns to simply analysis
df2 = df2.drop('annual_inc_joint', 1)
df2 = df2.drop('dti_joint', 1)
df2 = df2.drop('Individual_app', 1)
```
