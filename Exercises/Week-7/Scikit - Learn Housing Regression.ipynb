```python
X = sac[['zip', 'beds', 'baths', 'sq__ft', 'latitude', 'longitude']]
# creating a list of values for train_test_split function used below
```

```python
y = sac['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4)
```

```python
from sklearn import linear_model as lm
```

```python
LinReg = lm.LinearRegression()
```

```python
LinReg.fit(X_train,y_train)
```

```python
print(LinReg.intercept_)
```

```python
LinReg.coef_
```

```python
X_train.columns
```

```python
cLinReg = pd.DataFrame(LinReg.coef_, X.columns, columns=['Coeff'])
```

```python
cLinReg
```

```python
predictions = LinReg.predict(X_test)
#generates predicated values based off of Linear Regression Model using
#X_test values
```

```python
plt.scatter(y_test,predictions)
#scatterplot of actualy values versus predicted values
```

^^Doesn't look like our model worked very well with test data. No clear corrlation here between real values and predicted values

```python
sns.distplot((y_test-predictions))
#histogram of residuals. Linear Regression model may not be best fit.
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
